#Conhecendo a biblioteca Matplotlib


##Conhecendo os dados

Importação dos dados

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv")
df.head()
```

```python
df.info()
```

Realização de tratamentos

```python
df.set_index('País', inplace = True)
```

```python
df
```

```python
anos = list(map(str, range(1980,2014)))
```

```python
anos
```

Extração da série de dados para o Brasil

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(),'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

## Criando  o primeiro gráfico com Matplotlib

Criação de gráfico com módulo [pyplot](https://matplotlib.org/stable/tutorials/introductory/pyplot.html)

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

Melhora na visualização do gráfico

Escolhendo quais anos aparecerem no eixo x com [plt.xticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html)

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
```

Alterando as marcações do eixo y com [plt.yticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html)

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], color = 'g')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
```

Exibindo gráfico com [plt.show](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html)

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], color = 'g')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

Modificando o tamanho do gráfico com a função [plt.figure](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html)

```python
plt.figure(figsize = (8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], color = 'g')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

## Adicionando títulos e rótulos

Adicionando título com [plt.title](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.title.html)

```python
plt.figure(figsize = (8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], color = 'g')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Imigração do Brasil para o Canadá', fontsize = 15, pad = 10)
plt.show()
```

Adicionando rótulos aos eixos com [plt.xlabel](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.xlabel.html) e [plt.ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html)

```python
plt.figure(figsize = (8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], color = 'g')
plt.title('Imigração do Brasil para o Canadá', fontsize = 15, pad = 10)
plt.xlabel('Ano', fontsize = 10)
plt.ylabel('Número de imigrantes', fontsize=10)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

# Customizando com Matplotlib

## Personalizando visualizações com Matplotlib


```python
with plt.style.context('fivethirtyeight'):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, marker = 'o')
  ax.set_title('Imigração do Brasil para o Canadá\n1988 a 2013', fontsize = 18)
  ax.set_xlabel('Ano', fontsize = 14)
  ax.set_ylabel('Número de imigrantes', fontsize = 14)
  ax.grid(True, linestyle = '--')
  ax.xaxis.set_tick_params(labelsize = 12)
  ax.yaxis.set_tick_params(labelsize = 12)
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  plt.show()
```

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

with plt.style.context('fivethirtyeight'):
  fig, ax = plt.subplots(figsize=(12, 5))
  ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
  ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
  ax.set_xlabel('Número de imigrantes', fontsize=14)
  ax.set_ylabel('')
  ax.yaxis.set_tick_params(labelsize=12)
  ax.xaxis.set_tick_params(labelsize=12)

  plt.show()
```

## Alterando as cores com paletas

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize = (8,4))
  ax = sns.barplot(data = top_10, y = top_10.index, x='Total', orient = 'h', palette = palette)
  ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc = 'left', fontsize = 16)
  ax.set_xlabel("Número de Imigrantes", fontsize = 14)
  ax.set_ylabel('', fontsize = 14)

  plt.show()
```
