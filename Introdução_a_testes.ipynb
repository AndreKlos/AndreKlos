##Visualizando e limpando o movielens

```python
#carregando o arquivo ratings.csv

notas = pd.read_csv('/content/drive/MyDrive/Alura/Testes estatísticos com Python/data-science-introducao-a-testes-estatisticos-com-python-aula0/aula0/ratings.csv')
notas.head()
```

```python
#agrupando os filmes por nota média

nota_media_por_filme = notas.groupby('movieId').mean()['rating']
nota_media_por_filme.head()
```

```python
#criando a visualização dos dados
ax = sns.distplot(nota_media_por_filme.values)

ax.set(xlabel = 'Nota Média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no MovieLens')
```

```python
#gerando uma nova visualização com os filmes de pelo menos 10 votos

ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos)
ax.set(xlabel = 'Nota Média', ylabel = 'Densidade')
ax.set_title('Media de votos em filmes no MovieLens')
```

```python
#gerando a visualização em boxplot

ax = sns.boxplot(x = nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel = 'Nota média do filme')
ax.set_title('Distribuição de notas média dos filmes no MovieLens')
```

##Comparação não paramétrica com Wilcoxon Ranksums

```python
#utliziando teste para verificar que nossa distribuição é uma normal

from scipy.stats import normaltest

stats, p = normaltest(notas1['rating'])
p
```

não possuimos uma distribuição normal


```python
#pequena dica: para valores que não serão utilizados no restante do código, podemos atribuilo a _

_, p  = normaltest(notas1['rating'])

p
```

```python
#comparando os conjuntos separados anteriormente

from scipy.stats import ranksums

_, p = ranksums(notas1['rating'], notas593['rating'])
p
```

Pegando amostras aleatórias de cada conjunto, podemos descartar a hipotese nula de que as amostras vieram da mesma distribuição. Sendo assim, podemos afirmar que as notas de Silence of the Lambs (o segundo valor colocado) são, em média, maiores.
