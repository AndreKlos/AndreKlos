##Criando um banco de dados local

```python
#importando da biblioteca sqlalchemy

from sqlalchemy import create_engine, MetaData,Table, inspect

#criando o banco de dados usando o SQLite, nativo do Colab

engine = create_engine('sqlite:///:memory:')
```

##Escrevendo em um banco de dados

```python
#criando a variavel da url do github

url = "https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv"

#criando a variavel com os dados

dados = pd.read_csv(url)

#visualizando a parte inicial da tabela

dados.head()
```

```python
#colocando os dados em formato do banco, dando o novo nome da tabela

dados.to_sql('clientes', engine, index = False)
```

```python
#criando uma variavel para inspencionar o Banco de Dados

inspector = inspect(engine)

#imprimindo o nome das tabelas no banco de dados

print(inspector.get_table_names())

```

##Lendo uma consulta SQL

```python
#criando uma variavel para fazer a consulta

query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'

#criando a variavel para leitura e visualizando o resultado

empregados = pd.read_sql_query(query, engine)
empregados
```

```python
#salvando a nova tabela feita através da query

empregados.to_sql('empregados', con = engine, index = False)

```

```python
#visualizando a nova tabela criada

pd.read_sql_table('empregados', engine)
```

```python
#especificando qual colunas que gostariamos de consultar

pd.read_sql_table('empregados', engine, columns = ['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

##Atualizando um banco de dados

```python
#criando variavel para query

query='SELECT * FROM clientes'

#visualizando a pesquisa

pd.read_sql(query, engine)
```

Atualizando um banco de dados

```python
#fazendo o cancelamento do cliente ID 5008804

query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'

with engine.connect() as conn:
  conn.execute(query)
```

```python
#conferindo se o cliente foi excluido

pd.read_sql('clientes', engine)
```

Atualizando um dado

```python
#fazendo a atualização do grau de escolaridade do cliente 5008808

query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(query)
```

```python
#fazendo a consulta da modificação
pd.read_sql('clientes', engine)
```
